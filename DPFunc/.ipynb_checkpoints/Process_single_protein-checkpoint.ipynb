# DPFunc: single-protein preprocessing (minimal)
This notebook reduces the original `Process_data.ipynb` pipeline to **one protein**.

It generates the files DPFunc prediction needs:
- `processed_file/<ont>_test_used_pid_list.pkl`
- `processed_file/pdb_points.pkl`, `processed_file/pdb_seqs.pkl`
- `processed_file/protein_map.pkl`
- `processed_file/esm_emds/esm_part_0.pkl` (residue embeddings)
- `processed_file/graph_features/<ont>_test_whole_pdb_part0.pkl`
- `processed_file/interpro/<PID>.pkl` (zero-vector fallback unless you provide InterProScan-derived features)

After this, run:
```bash
python DPFunc_pred.py -d <ont> -n 0 -p <checkpoint_prefix>
```

Notes:
- The original DPFunc model expects InterPro features. If you do not provide them, this notebook writes a **zero vector** so the pipeline runs, but predictions may degrade.
- The residue-embedding step uses `esm` if available. If not, it falls back to calling `process_esm.py` (if your repo has it) or errors.


````python
import os, pickle as pkl, gzip, math
import numpy as np
from pathlib import Path
from tqdm.auto import tqdm

def read_pkl(path):
    with open(path, "rb") as f:
        return pkl.load(f)

def save_pkl(path, obj):
    Path(path).parent.mkdir(parents=True, exist_ok=True)
    with open(path, "wb") as f:
        pkl.dump(obj, f)
````

## 0) Set your inputs

````python
# EDIT THESE
PID = "A0S864"                   # a short ID used in filenames
ONTOLOGY = "mf"                    # "mf", "bp", or "cc"
PDB_PATH = f"./data/PDB/PDB_folder/{A0S864}.pdb"   # path to your pdb (can be relative)

# If you want to force CPU in graph building, set this True (prediction script still may need edits)
FORCE_CPU = False
````

## 1) Create folder structure + single-protein pid list

````python
Path("./processed_file").mkdir(exist_ok=True)
Path("./processed_file/graph_features").mkdir(parents=True, exist_ok=True)
Path("./processed_file/esm_emds").mkdir(parents=True, exist_ok=True)
Path("./processed_file/interpro").mkdir(parents=True, exist_ok=True)

save_pkl(f"./processed_file/{ONTOLOGY}_test_used_pid_list.pkl", [PID])
print("Wrote:", f"./processed_file/{ONTOLOGY}_test_used_pid_list.pkl")
````

## 2) Extract CA coordinates + sequence from your PDB

````python
from Bio.PDB import PDBParser
from Bio.SeqUtils import seq1

def extract_sequence_and_ca_coords(pdb_file, chain_id=None):
    parser = PDBParser(QUIET=True)

    # Support gz PDBs
    if str(pdb_file).endswith(".gz"):
        with gzip.open(pdb_file, "rt") as gz_file:
            tmp = str(pdb_file).replace(".gz", "_tmp.pdb")
            with open(tmp, "w") as out:
                out.write(gz_file.read())
        structure = parser.get_structure("protein", tmp)
        os.remove(tmp)
    else:
        structure = parser.get_structure("protein", pdb_file)

    model = structure[0]

    # Choose chain
    if chain_id is None:
        chains = list(model.get_chains())
        if not chains:
            raise ValueError("No chains found in PDB")
        chain = chains[0]
    else:
        chain = model[chain_id]

    seq = []
    ca_coords = []
    for residue in chain:
        # standard residues only
        if residue.id[0] != " ":
            continue
        if "CA" not in residue:
            continue
        try:
            aa = seq1(residue.resname)
        except Exception:
            continue
        seq.append(aa)
        ca_coords.append(residue["CA"].coord.astype(float))

    if not seq or not ca_coords:
        raise ValueError("Failed to extract sequence/CA coords (check chain selection and PDB completeness).")

    return "".join(seq), np.vstack(ca_coords)

seq, coords = extract_sequence_and_ca_coords(PDB_PATH, chain_id=None)
print("PID:", PID)
print("Sequence length:", len(seq))
print("CA coords shape:", coords.shape)

pdb_points_info = {PID: coords}
pdb_seqs = {PID: seq}

save_pkl("./processed_file/pdb_points.pkl", pdb_points_info)
save_pkl("./processed_file/pdb_seqs.pkl", pdb_seqs)
save_pkl("./processed_file/unseen_proteins.pkl", set())
print("Wrote processed_file/pdb_points.pkl, pdb_seqs.pkl")
````

## 3) Residue embeddings (ESM) for the single protein
This produces `processed_file/esm_emds/esm_part_0.pkl` and `processed_file/protein_map.pkl`.

````python
# Map protein -> embedding-part index (always 0 for single protein)
save_pkl("./processed_file/protein_map.pkl", {PID: 0})

# Try to embed with esm if available. If not, you must run the repo's process_esm.py (if present).
def embed_with_esm(seq: str):
    import torch
    import esm

    # Small model by default to reduce friction; change if you want
    model, alphabet = esm.pretrained.esm2_t12_35M_UR50D()
    model.eval()

    batch_converter = alphabet.get_batch_converter()
    data = [("protein", seq)]
    _, _, toks = batch_converter(data)

    with torch.no_grad():
        out = model(toks, repr_layers=[model.num_layers], return_contacts=False)
    rep = out["representations"][model.num_layers][0, 1:1+len(seq)]  # [L, C]
    return rep.cpu().numpy().astype(np.float32)

esm_out_path = "./processed_file/esm_emds/esm_part_0.pkl"

try:
    embs = {PID: embed_with_esm(seq)}
    save_pkl(esm_out_path, embs)
    print("Wrote:", esm_out_path, "shape:", embs[PID].shape)
except Exception as e:
    print("ESM embedding via `esm` failed.")
    print("Error:", repr(e))
    print("")
    print("Fallback:")
    print("1) Check whether your repo contains `process_esm.py`")
    print("2) If it does, run in a terminal:")
    print("   python process_esm.py")
    print("and ensure it writes `./processed_file/esm_emds/esm_part_0.pkl` containing a dict {PID: [L, C] numpy array}.")
    raise
````

## 4) Build the DGL graph + save graph_features pickle

````python
import torch
import dgl

def get_dis(p1, p2):
    d = p1 - p2
    return float(np.sqrt((d*d).sum()))

def build_graph_from_points(points: np.ndarray, threshold: float = 12.0):
    # O(L^2) edge building; fine for single proteins up to a few thousand residues.
    u_list, v_list, dis_list = [], [], []
    L = points.shape[0]
    for u in range(L):
        pu = points[u]
        for v in range(L):
            if u == v:
                continue
            dist = get_dis(pu, points[v])
            if dist <= threshold:
                u_list.append(u); v_list.append(v); dis_list.append(dist)

    u = torch.tensor(u_list, dtype=torch.int64)
    v = torch.tensor(v_list, dtype=torch.int64)
    dis = torch.tensor(dis_list, dtype=torch.float32)

    g = dgl.graph((u, v), num_nodes=L)
    g.edata["dis"] = dis
    return g

pdb_points_info = read_pkl("./processed_file/pdb_points.pkl")
protein_map = read_pkl("./processed_file/protein_map.pkl")
esm_features = read_pkl("./processed_file/esm_emds/esm_part_0.pkl")

points = pdb_points_info[PID]
g = build_graph_from_points(points, threshold=12.0)

node_features = esm_features[PID]
assert node_features.shape[0] == g.num_nodes(), (node_features.shape, g.num_nodes())

g.ndata["x"] = torch.from_numpy(node_features)

out_graph_path = f"./processed_file/graph_features/{ONTOLOGY}_test_whole_pdb_part0.pkl"
save_pkl(out_graph_path, [g])
print("Wrote:", out_graph_path)
````

## 5) InterPro features (zero-vector fallback)
DPFunc expects per-protein InterPro vectors. If you do not have InterProScan-derived features, this writes a zero vector using the repo’s `interpro_list_26203.pkl` (if present).

````python
interpro_list_path = "./data_dpfunc/interpro_list_26203.pkl"
if os.path.exists(interpro_list_path):
    interpro_list = read_pkl(interpro_list_path)
    vec = np.zeros(len(interpro_list), dtype=np.float32)
    save_pkl(f"./processed_file/interpro/{PID}.pkl", vec)
    print("Wrote zero-vector InterPro:", f"./processed_file/interpro/{PID}.pkl", "len:", len(vec))
else:
    # Minimal fallback: still write something (DPFunc may crash later if it strictly requires exact length).
    vec = np.zeros(1, dtype=np.float32)
    save_pkl(f"./processed_file/interpro/{PID}.pkl", vec)
    print("WARNING: interpro list not found at", interpro_list_path)
    print("Wrote placeholder InterPro vector of length 1. DPFunc may require the real length.")
````
