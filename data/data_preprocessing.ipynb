```python
import random
import pandas as pd

random.seed(42)

def read_fasta(filepath):
    sequences = []
    with open(filepath, "r") as f:
        seq = []
        for line in f:
            line = line.strip()
            if line.startswith(">"):
                if seq:
                    sequences.append("".join(seq))
                    seq = []
            else:
                seq.append(line)
        if seq:
            sequences.append("".join(seq))
    return sequences

# Load toxins (label = 1)
toxins = read_fasta("./Toxins.fasta")
n_toxins = len(toxins)

# Load non-toxins (label = 0)
non_toxins = read_fasta("./Non-toxins.fasta")

if len(non_toxins) < n_toxins:
    raise ValueError("Not enough non-toxin sequences to match toxin count.")

# Random subset of non-toxins equal to toxin count
non_toxins_subset = random.sample(non_toxins, n_toxins)

# Create DataFrame
data = pd.DataFrame({
    "sequence": toxins + non_toxins_subset,
    "toxicity": [1]*n_toxins + [0]*n_toxins
})

# Shuffle dataset
data = data.sample(frac=1, random_state=42).reset_index(drop=True)

print(data.head())
print(f"\nTotal samples: {len(data)}")
```
